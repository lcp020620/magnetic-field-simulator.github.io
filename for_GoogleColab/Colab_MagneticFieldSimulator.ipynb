```python
import cupy as cp
import numpy as np
import pandas as pd
from google.colab import files

from currentgenerator import CurrentGen
from magneticfieldsimulator import MagneticFieldSimulator
from vectorgenerator import VectorGenerator
from vectorgenerator import row
```

```python
simulator = MagneticFieldSimulator()
#=======================================================
#======Modify simulation space configurations here======
#======이 변수들을 조정해 시뮬레이션 공간을 조정하세요======
meshDense = 30
width = 10  #X axis | X 축
length = 10 #Y axis | Y 축
height = 10 #Z axis | Z 축
#=======================================================
r = simulator.makeMesh(dense=meshDense, width=width, length=length, height=height)
```

```python
current = CurrentGen()
#======Insert your current datas here======
#======여기에 전류 데이터를 생성하세요=======
#how to use "CurrentGen" class
# current.getLineCurrent(start=row(0, 0, 0), end=row(width, length, height), I=1, dense=20)
# current.getCircleCurrent(origin=row(0, 0, 0), face=row(1, 0, 0), radius=8, I=1 dense=30)
current.getCircleCurrent(origin=row(4, 0, 0), face=row(1, 0, 0), radius=8, I=1, dense=30)    #example of Helmholtz coil | 예제) 헬름홀츠 코일
current.getCircleCurrent(origin=row(-4, 0, 0), face=row(1, 0, 0), radius=8, I=1, dense=30)   #example of Helmholtz coil | 예제) 헬름홀츠 코일
#==========================================
dLArray = current.currentArr
```

```python
MF = simulator.makeMF(dLArray=dLArray)
plotMF = cp.asnumpy(MF)
plotr = cp.asnumpy(r)
csvArray = np.concatenate([plotr[:, 0:3], plotMF[:, 0:3]], axis=1)
pdcsvArray = pd.DataFrame(csvArray, columns=['x', 'y', 'z', 'vx', 'vy', 'vz'])
pdcsvArray.to_csv("MFArray.csv", index=False)
files.download("MFArray.csv")
```
